```python
import pandas as pd

info_bv = pd.read_csv('/home/mettalbi/projects/smash-maintenance-1.0.x/smash/factory/dataset/Ardeche/info_bv.csv')
aval = pd.read_csv('/home/mettalbi/projects/smash-maintenance-1.0.x/smash/factory/dataset/Ardeche/code_aval_ardeche.csv')
indep = pd.read_csv('/home/mettalbi/projects/smash-maintenance-1.0.x/smash/factory/dataset/Ardeche/code_indep_ardeche.csv')
```

```python
info_bv['Code_BV']
```

```python
indep.rename(columns={'val': 'code'}, inplace=True)
indep['label'] = 'indep'

aval.rename(columns={'val': 'code'}, inplace=True)
aval['label'] = 'aval'
```

```python
info_seg = pd.concat([indep, aval])
```

```python
amont = info_bv[~info_bv['Code_BV'].isin(info_seg['code'])][['Code_BV']]
amont.rename(columns={'Code_BV': 'code'}, inplace=True)
amont['label'] = 'amont'


info_seg = pd.concat([info_seg, amont])
```

```python
info_seg.to_csv('/home/mettalbi/projects/smash-maintenance-1.0.x/smash/factory/dataset/Ardeche/code_seg.csv', index=False)
```
